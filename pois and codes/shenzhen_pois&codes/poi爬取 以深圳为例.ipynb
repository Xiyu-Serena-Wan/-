```python
import xlwt
import xlrd
import requests
import json
import time
from IPython import embed
key = 'a1cf852d9781d1e427a06ab57400ce13'
url_ex = 'https://restapi.amap.com/v3/place/polygon?'
types = "010000"
data = xlrd.open_workbook("左下右上深圳.xls")
table = data.sheet_by_index(0)

n_rows = table.nrows #n_rows = 142
#table.row_values(0)=['113.198506999999, 23.372987| 113.278506999999, 23.452987']
wb = xlwt.Workbook()
ws = wb.add_sheet('深圳',cell_overwrite_ok=True)
```

```python
for i in range(123,n_rows):
    data_list = table.row_values(i)#在这个格子里
    lng_lat = data_list[0]
    url_lng_lat = url_ex + 'polygon=' + lng_lat + '&key=' + key + '&types=' + types + '&output=json&offset=20&extensions=base'
    print('第%d个网格数据 '%i)
    
    #进行翻页
    for p in range(1,50):
        url = url_lng_lat + '&page=' + str(p)
        html = requests.get(url)
        content = json.loads(html.text)

        list_name = []
        list_type = []
        list_pname = []
        list_cityname = []
        list_adname = []
        list_location = []
        
        for a in range(len(content['pois'])):#['name']
            #len(content['pois']) = 20
            if content['pois'][a]['name']  == 0:
                continue
            list_name.append(content['pois'][a]['name'])
            list_type.append(content['pois'][a]['type'])
            list_pname.append(content['pois'][a]['pname'])
            list_cityname.append(content['pois'][a]['cityname'])
            list_adname.append(content['pois'][a]['adname'])
            list_location.append(content['pois'][a]['location'])
            
        if len(list_name) == 0:
            continue
        print(p)
        for b in range(len(content['pois'])):
            ws.write((p-1)*20+b+i*50,0, list_name[b])
            ws.write((p-1)*20+b+i*50,1, list_type[b])
            ws.write((p-1)*20+b+i*50,2, list_pname[b])
            ws.write((p-1)*20+b+i*50,3, list_cityname[b])
            ws.write((p-1)*20+b+i*50,4, list_adname[b])
            ws.write((p-1)*20+b+i*50,5, list_location[b])
```

```python
wb.save('shenzhen_010000.xls')
```
